```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# file paths

labour = (r"NHS Scotland Workforce rates.xlsx")

wait_times =(r"2023-01-10-ed-weekly-nhsscotland-data.csv")

absence = (r"absencerates.xlsx")
```

```python
# read in all files

# workforce information
lab = pd.read_excel(labour)

# wait time information
wait = pd.read_csv(wait_times)

# absence numbers
abs_no = pd.read_excel(absence)
```

## Cleaning and sorting data

```python
lab.head()
```

```python
lab.describe()
```

```python
# melt data so dates are in one column

lab_melt= pd.melt(lab, id_vars=['Region'], var_name='date', value_name='workforce')
lab_melt.sort_values(by = ['Region'], inplace = True)
lab_melt.reset_index(drop = True, inplace = True)
lab_melt.head()
```

```python
## total all regions as do not require that level of granularity

labgrp = lab_melt .groupby(['date'])['workforce'].sum().reset_index()
labgrp.head()
```

```python
print(labgrp.shape)
print(labgrp.dtypes)
```

```python
# dates are running quarterly so add in year and quarter columns

labgrp['Year'] = labgrp['date'].dt.year
labgrp['Quarter_Year'] = labgrp['date'].dt.to_period('Q')
labgrp.head()
```

```python
# initial visualising of workfoce over time
x= labgrp['date']
plt.figure (figsize =(15,8))
sns.lineplot(x='date', y='workforce', data=labgrp)
plt.xticks(x, rotation = 90)
plt.xlabel('Date')
plt.ylabel ('No. of Employees')
plt.title('Total Number of NHS Employees')
plt.show()
```

```python
# read in absence info

abs_no = pd.read_excel(absence)
abs_no.head()
```

```python
# as with workforce need to melt table so dates are in a column

abs_melt= pd.melt(abs_no, id_vars=['Region'], var_name='date', value_name='No_of_absence')
abs_melt.sort_values(by = ['Region'], inplace = True)
abs_melt.reset_index(drop = True, inplace = True)
abs_melt.head()
```

```python
# don't require breakdown so group to total by each day
abs_grp = abs_melt .groupby(['date'])['No_of_absence'].sum().reset_index()

abs_grp.head()
```

```python
print(abs_grp.shape)
print(abs_grp.dtypes)
```

```python
# as with workforce information runs quarterly. Adding in year and quarter coloums for easier comaprisons

abs_grp['Year'] = abs_grp['date'].dt.year
abs_grp['Quarter_Year'] = abs_grp['date'].dt.to_period('Q')
abs_grp.head()
```

```python
#initial visulisation to get idea of data

x= abs_grp['date']
plt.figure (figsize =(15,8))
sns.lineplot(x='date', y='No_of_absence', data=abs_grp)
plt.xticks(x, rotation = 90)
plt.xlabel('Date')
plt.ylabel ('No. of Absences')
plt.title('Total Number of NHS Employees Absent')
plt.show()
```

```python
# read in A&E waiting times
wait = pd.read_csv(wait_times)
wait.head()
```

```python
#remove coloums not required
wait = wait[['week_ending_date','attendance', 'number_over_4_hours',
       'number_over_8_hours',
      'number_over_12_hours',
       ]]

#view table
wait.head()
```

```python
#check information
print(wait.shape)
print(wait.dtypes)
```

```python
wait.describe()
```

```python
#any missing values
wait.info()
```

```python
#change date to datetime instead of object for time series
wait['week_ending_date'] = pd.to_datetime(wait['week_ending_date'])

wait.dtypes
```

```python
# visulise the attendance over time
plt.figure (figsize =(15,8))
sns.lineplot(x='week_ending_date', y='attendance', data=wait)
plt.xticks(rotation = 90)
plt.xlabel('Date')
plt.ylabel ('No. of Patients')
plt.title('Total Number of People Attending A&E')
plt.show()
```

```python
#plot wait times

plt.figure (figsize =(15,8))
wait.plot(x="week_ending_date", y=['number_over_4_hours',
        'number_over_8_hours',
        'number_over_12_hours'],
        kind="line", figsize=(10, 10))

plt.xlabel('Date')
plt.ylabel ('No. of Patients')
plt.title('Wait Time for A&E Patients')
plt.show()
```

```python
# iformation weekly, so adding year and quarter coloums to allow information to be groped quarterly

wait['Year'] = wait['week_ending_date'].dt.year
wait['Quarter_Year'] = wait['week_ending_date'].dt.to_period('Q')
wait.head()
```

```python
waitgrp = wait.groupby(['Quarter_Year'])[['attendance', 'number_over_4_hours', 'number_over_8_hours', 'number_over_12_hours']].sum().reset_index()
waitgrp.head()
```

```python
# find date ranges for all datasets

print('attendance date min', waitgrp['Quarter_Year'].min())
print('attendance date max', waitgrp['Quarter_Year'].max())


print('workforce date min', labgrp['Quarter_Year'].min())
print('workforce date max', labgrp['Quarter_Year'].max())

print('absence date min', abs_grp['Quarter_Year'].min())
print('absence date max', abs_grp['Quarter_Year'].max())
```

```python
#adjusting dataframes so all cover same time period
waitgrp = waitgrp[(waitgrp['Quarter_Year'] >= '2015Q1' )& (waitgrp['Quarter_Year'] <= '2022Q3') ]
waitgrp.head()
```

```python
labgrp =  labgrp[( labgrp['Quarter_Year'] >= '2015Q1' )& (labgrp['Quarter_Year'] <= '2022Q3')  ]
labgrp.head()
```

```python
abs_grp =  abs_grp [(abs_grp ['Quarter_Year'] >= '2015Q1' )& (abs_grp['Quarter_Year'] <= '2022Q3')]
abs_grp .head()
```

```python
#create dataframe with workforce and absence coloums to wait times and attendance by quarter

NHS_times = pd.merge(waitgrp, labgrp, how = 'left', on = ['Quarter_Year'])
              
NHS_times = NHS_times[['Quarter_Year','Year', 'attendance', 'number_over_4_hours',
       'number_over_8_hours', 'number_over_12_hours', 'workforce']]


NHS_times = pd.merge(NHS_times ,abs_grp , how = 'left', on = ['Quarter_Year'])
NHS_times = NHS_times[['Quarter_Year','Year_x', 'attendance', 'number_over_4_hours',
       'number_over_8_hours', 'number_over_12_hours', 'workforce', 'No_of_absence']]

NHS_times.head()
```

```python
# Calculate total available workforce

NHS_times['adj_staff'] = NHS_times['workforce'] - NHS_times['No_of_absence']
NHS_times.head()
```

```python
# Change quater data types to string from period to allow analysis

NHS_times['Quarter_Year'] = NHS_times['Quarter_Year'].astype(str)
NHS_times.dtypes

```

## Analysis

```python
# What are attendance records for A&E?
NHS_times.describe()

```

```python
#compare similar time perids for 2019 and 2022 (miss 2020 and 21 due to COVID)

nhs_19_20 =  NHS_times[( NHS_times['Quarter_Year'] == '2019Q3') | (NHS_times['Quarter_Year'] == '2022Q3')]

#what is the % change for these 2 figures

nhs_19_20['att_%change'] = nhs_19_20['attendance'].pct_change()
nhs_19_20

```

```python
plt.figure (figsize =(15,8))
sns.lineplot(x='Quarter_Year', y='attendance', data=NHS_times)
plt.xticks(rotation = 90)
plt.xlabel('Date')
plt.ylabel ('No. of Patients')
plt.title('Number of People attending A&E')
plt.show()
```

```python
# plot how have wait times changed?

NHS_times.plot(x="Quarter_Year", y=['number_over_4_hours',
        'number_over_8_hours',
        'number_over_12_hours'],
        kind="line", figsize=(15, 10))
plt.xlabel('Date')
plt.ylabel ('No. of Patients')
plt.title('Wait Time for A&E Patients')

plt.show()
```

```python
#what is the % change for these wait times for Q3 2019 and 2022

nhs_19_20['4_%change'] = nhs_19_20['number_over_4_hours'].pct_change()
nhs_19_20['8_%change'] = nhs_19_20['number_over_8_hours'].pct_change()
nhs_19_20['12_%change'] = nhs_19_20['number_over_12_hours'].pct_change()
nhs_19_20 [['Quarter_Year', 'number_over_4_hours', '4_%change', 'number_over_8_hours', '8_%change', 'number_over_12_hours', '12_%change']]
```

```python
# plot workforce nubers and absence numbers togther see if they follow each other

NHS_times.plot(x="Quarter_Year", y=['workforce',
        'No_of_absence'],
        kind="line", figsize=(10, 5))
plt.xlabel('Date')
plt.ylabel ('No of Employees')
plt.title('Numbers of Total Employees and Absent Employees')

plt.show()
```

```python
# calculate percentage chnage for workforce, abesence and adjusted workforce

nhs_19_20['%workforce'] = nhs_19_20['workforce'].pct_change()
nhs_19_20['%No_of_absence'] = nhs_19_20['No_of_absence'].pct_change()
nhs_19_20['%adj_staff'] = nhs_19_20['adj_staff'].pct_change()
nhs_19_20 [['Quarter_Year', 'workforce','%workforce', 'No_of_absence', '%No_of_absence', 'adj_staff','%adj_staff' ]]
```

```python
# plot adjusted workforce numbers

x= NHS_times['Quarter_Year']
plt.figure (figsize =(15,8))
sns.lineplot(x='Quarter_Year', y='adj_staff', data=NHS_times)
plt.xticks(x, rotation = 90)
plt.xlabel('Date')
plt.ylabel ('No. of Employees')
plt.title('Number of NHS Employees')
plt.show()
```

```python
#are workforce and wait times correlated
NHS_times[['number_over_4_hours','number_over_8_hours', 'number_over_12_hours', 'adj_staff' ]].corr()
```

```python
#visualise correlation
sns.heatmap(NHS_times[['number_over_4_hours','number_over_8_hours', 'number_over_12_hours', 'adj_staff']].corr(), annot=True)
```

```python
#plot wait times agaist workforce numbers

fig, ax1 = plt.subplots(figsize=(15, 8))
ax2 = ax1.twinx()

ax1.plot(NHS_times['Quarter_Year'], NHS_times['number_over_4_hours'], color = 'r')
ax2.plot(NHS_times['Quarter_Year'], NHS_times['adj_staff'], color='b')

ax1.set_xlabel("Date")
ax1.set_ylabel("No of Patients waiting over 4 hours", color='r', fontsize=14)
ax2.set_ylabel("Number of Available Employees", color='b', fontsize=14)

fig.suptitle("Increase in wait times compared to staff levels", fontsize=20)
ax1.set_xticklabels(NHS_times['Quarter_Year'], rotation = 90)
plt.show()
```
